```python
from pycocotools.coco import COCO
import json
import os
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
json_path = "/Users/rimo/Documents/paper/data/data.json"
data_csv = "/Users/rimo/Documents/paper/data/test_for_cam-bbox.csv"
coco = COCO(json_path)
data = pd.read_csv(data_csv)
```

```python
cnt = 120
```

```python
n = cnt 
print(n)
img_info = coco.loadImgs(n)[0]
print(img_info)
img = cv2.imread("/Users/rimo/Documents/paper/data/images/"+data["image_path"][n])

anns_ids = coco.getAnnIds(imgIds=data["image_id"][n], iscrowd=None)
anns = coco.loadAnns(anns_ids)
print(anns)
fig, axes = plt.subplots(1,1,figsize=(15,10))
for i in anns:
    bbox = i["bbox"]
    x1 = int(bbox[0])
    y1 = int(bbox[1])
    x2 = x1 + int(bbox[2])
    y2 = y1 + int(bbox[3])
    img = cv2.rectangle(img,(x1,y1),(x2,y2),[255,255,0],10)
# plt.imshow(img)
# cnt +=1
```
